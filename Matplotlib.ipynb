# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4,5]
y = [99,5,8,87,54]

#2. setup plot
fig, ax = plt.subplots(figsize = (10,10)) ## (width,height) specifyng height and width

# #.Plot data
ax.plot(x,y)

# 4.Customise plot
ax.set(title = "Simple plot",xlabel = "X-axis",ylabel = "Y-axis")

#5. Save and show whole fig
fig.savefig("C://Users//shrey//OneDrive//Desktop//imagesofgraph.png")
```

## Two options for subplots

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
ax1.plot(x,x**3);
ax2.scatter(x,np.sqrt(x));
ax3.plot(x,np.cos((x**2)/180));
ax4.plot(x,np.log(x));
```

```python
# Subplots 2nd type
fig,ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
# plot to each dfferent index
ax[0,0].barh(x,x/2)
ax[0,1].plot(np.random.randint(0,100,3),np.sin(np.random.randint(0,100,3)))
ax[1,1].scatter(x,x**2)
ax[1,0].plot(np.random.randint(0,100,6),np.tan(np.random.randint(0,100,6)))
```

 # pyplot or matplotlib  method : 
 * if something to quick then pyplot
 * if advanced plottin then oo

```python
over_50 = df[df.age>50]
```

```python
over_50.head()
```

```python
# PYPLOT 
over_50.plot(kind = "scatter",x = "age",y = "chol", c= "target");
```

```python
# Here is c for target 
```

```python
fig,ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter", x = "age",y = "chol",c = "target",ax = ax);
# we can set axis size
ax.set_xlim([45,100]) # sets the x axis limit
```

```python
# oo mathod from scratch pure
fig,ax = fig,ax = plt.subplots(figsize = (10,6))

# plot the data
scatter = ax.scatter(x = over_50.age,y=over_50.chol,c = over_50.target)
# customise
ax.set(title = "Heart disease and cholesterol level", xlabel = "Age",ylabel = "Cholesterol");
ax.legend(*scatter.legend_elements(), title = "Target");
# takes the data from scatter variable then find the legend elements i.e c parameter target and then unpaks legend elements and addd title
# add a horizontal line
ax.axhline(over_50.chol.mean(),linestyle = "--")
```

```python
 # subplot of chol,age,thalach
fig,(ax0,ax1) = plt.subplots(nrows = 2,ncols = 1,figsize = (10,10))
scaterr = ax0.scatter( x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target)
# Customise ax0
ax0.set(title = "Heart disease cholesterol",xlabel = "Age",ylabel = "cholesterol");
ax0.legend(*scatter.legend_elements(), title = "Target")
ax0.axhline(y = over_50.chol.mean(), linestyle= "--")

scatter2 = ax1.scatter(x = over_50.age,
                      y = over_50.thalach,
                      c = over_50.target)
ax1.set(title = "heart disease Thalach",xlabel = "age",ylabel = "THalach")
ax1.legend(*scatter2.legend_elements(),title = "Target")
ax1.axhline(y = over_50.thalach.mean(),linestyle = "--")
```
