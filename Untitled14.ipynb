```python
def plot_bar(state , pop , legend = "population" , title = 'State vs population'):
    plt.bar(state , pop , width = 0.5, color = "g" )
    plt.yticks(fontsize=12, color='purple') 
    plt.xticks(ticks = [0,1,2,3,4],labels = state,rotation = 45)
    plt.title(title, 
              fontsize=20,         
              fontweight='bold',   
              color='green',       
               
              style='italic',       
              pad=20)               
    plt.legend([legend])
    plt.figure(figsize = (10,10));
    
    

```

```python
firms = ["LIC","ICICI",'SBI','ABC','XVA']
tax = [23245,23455,45345,3454,66535]
plot_bar(firms,tax,"Tax","Firms vs tax collection")
```

```python
num_regions = 10 
min_houses = 50  
max_houses = 500  

houses_per_sq_km = np.random.randint(min_houses, max_houses + 1, size=num_regions)

print("Number of houses per square kilometer in each region:")
print(houses_per_sq_km)

plt.figure(figsize=(10, 6))
plt.barh(range(num_regions), houses_per_sq_km, color='skyblue', alpha=0.7)
plt.xlabel('Houses per sq km')
plt.ylabel('Region')
plt.title('Number of Houses per Square Kilometer in Different Regions')
plt.xticks(np.arange(min_houses, max_houses + 1, 50))
plt.legend(["Number of Houses"], loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

karnataka_density = np.array([365, 370, 380, 390, 400])
tamil_nadu_density = np.array([343, 350, 360, 370, 380])
kerala_density = np.array([455, 460, 470, 480, 490])
andhra_pradesh_density = np.array([454, 460, 470, 480, 490])
telangana_density = np.array([454, 460, 470, 480, 490])

states = ['Karnataka', "Tamil Nadu", "Kerala", "Andhra Pradesh", "Telangana"]
densities = [karnataka_density, tamil_nadu_density, kerala_density, andhra_pradesh_density, telangana_density]


fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()

for i in range(len(states)):
    axs[i].bar(np.arange(len(densities[i])), densities[i], color='skyblue', alpha=1.0)
    axs[i].set_title(states[i])
    axs[i].set_ylabel('Population Density (people per 1000 km)')
    axs[i].set_ylim(0, 500)
    axs[i].grid(axis='y')


axs[5].axis('off')

plt.tight_layout()
plt.suptitle('Population Density of Indian States', fontsize=16, y=1.02)
plt.show()
```
